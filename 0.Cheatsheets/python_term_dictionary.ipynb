### 1. Basic Terms

- **Program**          :
 A sequence of definitions and commands

- **Commands**         : 
Statements that instruct the interpreter to do something.

- **Static language**  : 
Language where the data types for each variable have to be specified so it can enforce data types. Java is an example of this.

- **Dynamic language** : 
Language where the data types for each variable DON'T have to be specified as it doesn't enforce data types. THIS IS WHAT PYTHON IS.

- **Module**           : 
A file that contains definitions—including variables and functions—that you can use once it is imported

- **Built-in Function** : 
A function that is built in to the programming language and so does not need to be defined or imported from a module to be called.

### 2. Types of Algorithm

- **Exhaustive Enumeration:** 

A set of algortihms using guess-and-check methods that can work on problems where there's a finite number of choices.

Is a good way to generate guesses in an organized manner using iterative loops as a technique to tell when you're done and to return a great answer.


                           
- **Guess-and-check:**  

An algorithm that using the process of exhaustive enumeration, gets you to:
1. Guess the value of a solution
2. Check if the solution is correct
3. Keep guessing until you find the correct solution or have exhausted all possible values
                              


### 3. General Syntax

- **Comment:**  
          
 Extra notes in your program that the computer ignores. Can be either a '#' or ' " ' for one line comments or ' """ ' for multiple line comments. 

    
Can use comments to :
                    
1. Explain parts of the program, for you or other people to read later;
2. Leave "to do" notes, when you write a longer program; [looks like for example # todo: check this]
3. Temporarily disable ("comment out") a line of a program without totally deleting it, so that it is easier to put back in later.
        

- **Indentation:** 

The white space used after a command such as an if statement to indicate a new block of text specified to that if statement - used the same way as the `{` in Java.


```python
if 8 < 9:
    print ("Eight is less than nine!")
```

If the indentation was not put before the print statement Python would get confused and you would get an error.
Likewise if you put a different indentation from one line to another (e.g no indentation to a two-space indentation or two-space indentation to four-space indentation) without a command like the if statement it will create an error.


- **Backslash notation:**

Looks like `\` . Is used to create a string that contains both single and double quotes without running into a syntas error

For example:
                        

```python
this_string = '"Simon\'s skateboard is in the garage." She said.' 
```

- **Concatenation:**

 Pulling more than one string together using a + operator

### 7. Data Structures 

- **Mutability:** 

Whether or not we can change an object once it has been created. 
             
If an object (like a list or string) can be changed (like a list can), then it is called mutable. 

e.g. 

```python
my_lst = [1, 2, 3, 4, 5]
my_lst[0] = 'one'
print(my_lst)
```

However, if an object cannot be changed with creating a completely new object (like strings), then the object is considered immutable.

```python
greeting = "Hello there"

# TypeError: 'str' object does not support item assignment
greeting[0] = 'M'
```

- **Ordered vs Unordered:**

Whether the position of an element in the object can be used to access the element.

- **List:** 

(`list` for short)

MUTABLE and ORDERED.

Stores a collection of different pieces of information as a sequence under a single variable name.

Denoted by *square brackets*

```python
my_list = ['my string', 0, True]
```

- **Tuple:**   

IMMUTABLE and ORDERED.

Denoted by circular brackets:

```python
my_tuple = (-3, 1)
```

Good for values that are so closely-relatd that they will always be used together e.g. longitude and latitutude
              
Also good for assigning multiple variables at once

```python
length, width, height = 52, 40, 100
```

- **Set:**

MUTABLE and UNORDERED collections of *unique* elements.

Denoted by curly brackets (however when assigning an *empty* set to a variable, you need to specify it's a set!)

```python
my_set = {1,2,3,4}
empty_set = set{}
```

Often used to quickly get rid of duplicates in a list!

```python
numbers = [1, 2, 6, 3, 1, 1, 6]
unique_nums = set(numbers)
print(unique_nums)
```

- **Dictionary:**

ORDERED and MUTABLE objects that store mappings of unique keys to values.

```python
elements = {"hydrogen": 1, "helium": 2, "carbon": 6}
emtpy_dict = {}
```

Keys can be of any data type as long as they are *immutable* (e.g. str, int, float, tuple etc. but NOT another List!)

Keys can be of different types within the SAME dictionary too!
                   
We can look up values or insert new values in the dictionary using square brackets that enclose the key.

```python
elements = {"hydrogen": 1, "helium": 2, "carbon": 6}

print(elements["helium"])  # print the value mapped to the "helium" key
elements["lithium"] = 3  # insert "lithium" with a value of 3 into the dictionary
```
